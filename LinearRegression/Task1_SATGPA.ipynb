```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# Change to data path on your computer
data =pd.read_csv("SAT_GPA.csv")
# Show the description of data
data.describe()
# Set to training data (x, y)
y = data['GPA']
x = data['SAT']
# Remind that we need to put component x_0 = 1 to x
plt.scatter(x,y)
plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPA', fontsize = 20)
plt.show()
```

Hãy chia dữ liệu thành phần training với 60 mẫu đầu và validation với các mẫu còn lại.

```python
training = data[0:60]
testing = data[60:data.shape[0]]

x_train = training["SAT"]
y_train = training["GPA"]

x_test = testing["SAT"]
y_test = testing["GPA"]
```

```python
from sklearn import datasets, linear_model
# Convert to numpy arrays
regr = linear_model.LinearRegression()
# fit_intercept = False for calculating the bias
regr.fit(x_train.values.reshape(-1, 1), y_train)
print(regr.coef_)
y_pred = regr.predict(x_train.values.reshape(-1, 1))
```

```python
plt.scatter(x_train, y_train)
fig = plt.plot(x_train,y_pred, lw=4, c='orange', label = 'regression line')
plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPA', fontsize = 20)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

# Make predictions on the validation set
y_test_pred = regr.predict(x_test.values.reshape(-1, 1))

# Calculate the mean squared error on the validation set
mse = mean_squared_error(y_test, y_test_pred)
print("Mean Squared Error:", mse)
```
